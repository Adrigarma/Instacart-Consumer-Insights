# 03. Grouping data and aggregating variables

## 03.2 Average orders by department

```python
# First split data into groups, then apply the aggregation function
ords_prods_merge.groupby('department_id').agg({'order_number':['mean']})
```

## 03.3 Analysis from results

## 03.4 Loyalty flag for customers

```python
# Split data based on user_id, apply transform() to generate the maximum orders for each user and create a new column "max_orders'
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge['max_order'].value_counts(dropna = False)
```

```python
ords_prods_merge.head(100)
```

```python
# Assign the loyalty lebel to the user_id
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge.groupby(['loyalty_flag']).user_id.nunique()
```

```python
# Checking new column
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

## 03.5 Difference between spending habits of the three types of customers

```python
# mean, min and max of prices grouped by loyalty_flag
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean', 'min', 'max']})
```

The basic statistics of the product prices for each type of customer is very similar. Minimum and maximum prices are the same for the three types of customer. While the lowest mean is for the loyal customer 7.77 , followed by the regular customer and the new customer with the highest mean of 7.8. Theres no a significant difference between the three types of customers. 
I will now look for the information that the price_flag gives.

```python
# look for the frecuency of the price_flag grouped by the loyalty_flag
ords_prods_merge.groupby('loyalty_flag')['price_range_loc'].value_counts(dropna = False)
```

The consume of products is similar for the three types of customers, being the mid-range products the most consumed among the three types of customer and the high-range products the least consumed. 

## 03.6 Spending flag

```python
# Split data based on user_id, and aggregate the mean of the price
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge['avg_price'].value_counts(dropna = False)
```

```python
# Asign spending flag to each user_id
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.groupby(['spending_flag']).user_id.nunique()
```

```python
ords_prods_merge[['user_id', 'spending_flag', 'avg_price']].head(60)
```

## 03.7 Frequent vs. non-frequent customers

```python
# Split data based on user_id, and aggregate the median of days_since_last_order
ords_prods_merge['days_last_order_median'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
ords_prods_merge['days_last_order_median'].value_counts(dropna = False)
```

```python
# Asign frequency flag to each user_id
ords_prods_merge.loc[ords_prods_merge['days_last_order_median'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['days_last_order_median'] > 10) & (ords_prods_merge['days_last_order_median'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['days_last_order_median'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.groupby(['frequency_flag']).user_id.nunique()
```
