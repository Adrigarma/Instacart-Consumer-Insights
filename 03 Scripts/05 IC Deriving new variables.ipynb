## 03. Deriving new variables

## 03.1 If statements with user-defined functions

```python
# Define the function 'def' followed by the name 'price_label'  
# The argument (row) #looks at each row within the dataframe and : separates the head from the body

def price_label(row):  #

    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
          return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
    else: return 'Not enough data'
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

## 04.4 Busiest hours of the day (periods of time)

```python
# Find out which hours of the day have the most orders 

ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

Looking at the values, I decided to divide the periods of time according to the amount of orders: Most_orders will be the hours with >1,000,000 orders (10,11,14,15,13,12,16,9,17,8,18,19), Average_orders for the hours with >100,000 and <1,000,000 (20,7,21,22,23,6,0,1) and Fewest_orders will be the hours with >100,000 orders (5,2,4,3). 

```python
# Create the statement for the new column

result_hours = []

# Define values group

group_most = [10,11,14,15,13,12,16,9,17,8,18,19]
group_average = [20,7,21,22,23,6,0,1]

# Define the loop function 

for value in ords_prods_merge['order_hour_of_day']:
    if value in group_most:
        result_hours.append('Most_orders')
    elif value in group_average:
        result_hours.append('Average_orders')
    else:
        result_hours.append('Fewest_orders')
```

```python
result_hours
```

```python
# Create column for result_days

ords_prods_merge['busiest_period_of_day'] = result_hours
```

```python
# Frequency of new column

ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
