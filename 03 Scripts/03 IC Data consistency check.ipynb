## 03.1 Mixed data types 

Example to deal with mixeddata types (Instacart has not mixed-type columns)

```python
# Create a dataframe
df_test = pd.DataFrame()
```

```python
# Create a mixed type column
df_test['mix'] = ['a', 'b', 1, True] 
```

```python
df_test.head()
```

```python
# function for checking whether a dataframe contains any mixed-type column
for col in df_test.columns.tolist(): # for loop (for this elements do this)
    weird = (df_test[[col]].applymap(type) != df_test [[col]].iloc[0].apply(type)).any(axis = 1) # variable weird with test that checks for 
#consistent datatypes within the columns. The weird variable take a boolean value of either True or False
# True (1) - column contains inconsistent data types
# False (0) - column contains only one data type
    if len (df_test[weird]) > 0: # if weird is true, then the command print is executed
        print(col)
```

```python
df_test['mix'] = df_test['mix'].astype('str')
```

Adressing missing values

Three ways with Python:
1. Create a new variable that acts like a flag based on the missing value.
2. Impute the value wiht the mean or median (only numeric values).
3. Remove or filter missing data.

```python
# creating a new dataframe that excludes de null values (because string values can't be imputed like numeric values)
df_prods_clean = df_prods[df_prods['product_name'].isnull() == False]
```

```python
df_prods_clean.shape
```
