# 04. Wrangling data 

## 04.1 Rename columns 

```python
# Renaming columns names to follow same format and logic
customers.rename(columns = {'First Name' : 'first_name', 'Surnam' : 'surname', 'Gender' : 'gender', 'STATE' : 'state', 'Age' : 'age', 'date_joined' : 'customer_since'}, inplace = True)
```

```python
customers.columns
```

## 04.2 Drop columns

I will drop first_name and surname as they are not useful for the analysis and also for privacy issues could be necessary. 

```python
# Create a new dataframe without the mentioned columns
customers_drop = customers.drop(['first_name', 'surname'], axis=1)
```

```python
customers_drop.head()
```

## 04.3 Change datatype

I will change the datatype of customer_since to datetime and age and n_dependants to int8

```python
# Change data type to datetime
customers_drop['customer_since'] = customers_drop['customer_since'].astype('datetime64[ns]')
```

```python
# Change datatype to int8
customers_drop[['age', 'n_dependants']] = customers_drop[['age', 'n_dependants']].astype('int8')
```

```python
customers_drop.dtypes
```

```python
customers_drop.head()
```

Now I can include the dates in the descriptive analysis

```python
customers_drop['customer_since'].describe()
```

# 05. Consistency check

## 05.1 Mixed data types

```python
# Check for mixed daty-types columns
for col in customers_drop.columns.tolist():
    weird = (customers_drop[[col]].applymap(type) != customers_drop[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (customers_drop[weird]) > 0:
        print(col)
```

There are no mixed-typed columns

# 06. Merge data 

```python
ords_prods_merge.dtypes
```

The key column is user_id, which in the customers dataset is type int64, and in the ords_prods_merge is ist32. Then I will change the dtype in the customers data. 

```python
# Change data type for column user_id in customers dataframe
customers_drop['user_id'] = customers_drop['user_id'].astype('int32')
```

```python
customers_drop.dtypes
```

```python
# Merge both dataframes
ords_prods_all = ords_prods_merge.merge(customers_drop, on = 'user_id', indicator = True)
```

```python
ords_prods_all.head()
```

```python
# Check if it's a full match
ords_prods_all['_merge'].value_counts()
```

```python
ords_prods_all.info()
```

```python
# Remove '_merge' columnb
ords_prods_all = ords_prods_all.drop(['_merge'], axis=1)
```
