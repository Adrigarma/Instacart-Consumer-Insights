Income segmentation

```python
# Check income distribution to designate income segmentation
ords_prods_all['income'].plot.hist(bins=50)
```

```python
# Create a column for income-segmentation
ords_prods_all.loc[ords_prods_all['income'] > 60000, 'income_segment'] = 'low income (< 60,000)' 
```

```python
ords_prods_all.loc[(ords_prods_all['income'] <= 60000) & (ords_prods_all['income'] < 180000), 'income_segment'] = 'mid income (60,000-179,000)' 
```

```python
ords_prods_all.loc[ords_prods_all['income'] >= 180000, 'income_segment'] = 'high income (180,000 +)'
```

```python
ords_prods_all['income_segment'].value_counts(dropna=False)
```

```python
# Check number of users in each income segment
ords_prods_all.groupby ('income_segment')['user_id'].nunique()
```

Customer profiles definition

```python
# Explore connections between the three different segmentations previously established and flags
segments_explor = ords_prods_all.groupby(['age_group', 'income_segment', 'family_profile'])['user_id'].nunique().reset_index().rename(columns={'user_id': 'customer_count'})
```

```python
segments_explor
```

```python
# Create subset
customer_profiling = ords_prods_all.groupby('user_id')[['age_group', 'income_segment', 'family_profile']].first().reset_index()
```

```python
customer_profiling.head()
```

```python
customer_profiling.shape
```

```python
# Create customer profile multi-gen families
customer_profiling.loc[(customer_profiling['age_group'] == 'young adult') & 
(customer_profiling['family_profile'] == 'Living with parents and siblings, with dependants'),'customer_profile'] = 'Multi-gen Families'
```

```python
# Create customer profile independent adults
customer_profiling.loc[
(customer_profiling['age_group'].isin(['young adult', 'middle-aged adult'])) &
 (customer_profiling['income_segment'].isin(['low income (< 60,000)', 'mid income (60,000-179,000)'])) &
(customer_profiling['family_profile'] == 'Single, no dependants'),'customer_profile'] = 'Independent Adult'
```

```python
# Create customer profile for Established families 
customer_profiling.loc[
    (customer_profiling['income_segment'] == 'mid income (60,000-179,000)') &
    (customer_profiling['family_profile'] == 'Married with dependants'), 'customer_profile'] = 'Established Families'
```

```python
# Low-Budget Families
customer_profiling.loc[
    (customer_profiling['income_segment'] == 'low income (< 60,000)') &
    (customer_profiling['family_profile'] == 'Married with dependants'), 'customer_profile'] = 'Low-Budget Families'
```

```python
# Retired and Solo
customer_profiling.loc[
    (customer_profiling['age_group'] == 'senior adult') &
    (customer_profiling['family_profile'] == 'Divorced/widowed, no dependants'),'customer_profile'] = 'Retired and Solo'
```

```python
# Affluent Customers
customer_profiling.loc[(customer_profiling['income_segment'] == 'high income (180,000 +)'), 'customer_profile'] = 'Affluent Customers' 
```

```python
prof_counts = customer_profiling['customer_profile'].value_counts(dropna=False).sort_values(ascending = True)
```

```python
prof_counts
```

Visualization

```python
profiles = prof_counts.plot.barh(
    color = ['springgreen', 'palegreen', 'limegreen', 'mediumseagreen', 'seagreen', 'forestgreen'], rot=0)
profiles.set_title('Customer Profiles')
profiles.set_xlabel('number of customers')
profiles.set_ylabel('')
plt.tight_layout()
```

```python
# Export barchart
profiles.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'customer_profiles_bar.png'))
```

```python
# Merge subset with dataframe
ords_prods_all_profiles = ords_prods_all.merge(customer_profiling[['user_id', 'customer_profile']], on = 'user_id', how='left')
```
